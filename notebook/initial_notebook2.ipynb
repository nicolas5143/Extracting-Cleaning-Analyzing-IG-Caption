```python
import ast

cats = list()
if 'comp_category' in data.columns:
    for cat_str in data['comp_category'].dropna():
        try:
            for c in ast.literal_eval(str(cat_str)): 
                if c not in cats:
                    cats.append(c)
        except: 
            pass

len(cats)
```

```python
def is_not_alphabet(text):
    for char in text:
        if char in ['-', ' ', '/', '&', ')', '(']:
            continue
        if not char.isalpha():
            return True  
    return False

for cat in cats:
    if is_not_alphabet(cat):
        print(cat)
```

```python
import ast

known_categories = list()
good_rows = data[data['post_category'] != 'Unclassified']

if 'comp_category' in good_rows.columns:
    for cat_str in good_rows['comp_category'].dropna():
        try:
            cat_list = ast.literal_eval(str(cat_str)) # making it to list
            for c in cat_list: 
                if c not in known_categories: known_categories.append(c)
        except: 
            pass

print(f"âœ… Loaded {len(known_categories)} categories into memory.")
```
