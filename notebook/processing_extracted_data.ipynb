### Normalizing the competition categories

```python
# making a function to cumulate the categories in the list
def cumulate_cats(datas):
    cats = list()
    for cats_list in datas.comp_category.dropna():
        if '[' not in str(cats_list):
            if cats_list not in cats:
                cats.append(cats_list)
            continue

        cats_list = ast.literal_eval(str(cats_list))
        for cat in cats_list:
            if cat not in cats:
                cats.append(cat)

    return cats

cats = cumulate_cats(data)
print(f'len: {len(cats)}')

# cats[:20]
```

## Target Audience

### Normalizing the target audience

```python
for ind in range(data.shape[0]):
    entry = data.iloc[ind]
    target_audiences = entry.target_audience

    if 'nan' in str(target_audiences):
        data['target_audience'].iloc[ind] = ['Invalid']
        continue
    
    auds_list = list()
    data_auds = ast.literal_eval(str(target_audiences))

    if len(data_auds) == 0:
        data['target_audience'].iloc[ind] = ['Invalid']
        continue

    for aud in data_auds:
        added = False

        # umum
        if (('umum' in aud.lower()) | ('non-ui' in aud.lower()) | ('publik' in aud.lower()) | ('alumni' in aud.lower()) | 
            ('guru' in aud.lower()) | ('pelaku' in aud.lower()) | ('gen-z' in aud.lower()) | ('dosen' in aud.lower())):
            auds_list.append('Umum')
            added = True
            continue
        # mahasiswa 
        if ('mahasiswa' in aud.lower()) | ('s1' in aud.lower()) | ('d4' in aud.lower()) | ('d3' in aud.lower()):
            auds_list.append('Mahasiswa')
            added = True
            continue
        # sma
        if (('sma' in aud.lower()) | ('pelajar' in aud.lower()) | ('derajat' in aud.lower()) | ('smk' in aud.lower()) | 
            ('siswa' in aud.lower()) | ('mts' in aud.lower()) | ('mi' in aud.lower()) | ('ma' in aud.lower())) & ('maha' not in aud.lower()):
            auds_list.append('SMA')
            added = True
            continue
    
        if not added:
            auds_list.append(aud)
    
    data['target_audience'].iloc[ind] = list(set(auds_list))
```

### The min and max for the registration fee 

```python
data['min_registration_fee'] = -1
data['max_registration_fee'] = -1

for ind in range(data.shape[0]):
    entry = data.iloc[ind]
    fees_list = ast.literal_eval(str(entry.registration_fee))

    if len(fees_list)==0:
        data['registration_fee'].iloc[ind] = [-1]
        fees_list = [-1]

    data['min_registration_fee'].iloc[ind] = min(fees_list)
    data['max_registration_fee'].iloc[ind] = max(fees_list)
```

### Stating if the registration fee is stated in the caption or not

```python
data['fee_type'] = 'Not listed'

for ind in range(data.shape[0]):
    entry = data.iloc[ind]
    fees_list = ast.literal_eval(str(entry.registration_fee))

    if len(fees_list)==1:
        if fees_list[0] == 0:
            data['fee_type'].iloc[ind] = 'Free'
            continue
        if fees_list[0] == -1:
            data['fee_type'].iloc[ind] = 'Not listed'
            continue
    
    data['fee_type'].iloc[ind] = 'Paid'
```
